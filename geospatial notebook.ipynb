## Geospatial

```sql
-- Returns all parking meters physically located in the Barrio Logan neighborhood.
SELECT
  pole,
  latitude,
  longitude,
  area AS neighborhood
FROM parking_catalog.parking_schema.parking_meter_man
WHERE area = 'Barrio Logan';
```

```sql
-- Returns all meters geographically located within the specified ward boundary.
SELECT
  pole,
  latitude,
  longitude,
  area
FROM parking_catalog.parking_schema.parking_meter_man
WHERE latitude BETWEEN 32.69 AND 32.73
  AND longitude BETWEEN -117.25 AND -117.14;
```

## INNER JOIN

```sql

--Returning only the parking meters that have matching entries in the Rate table, showing their zone, area, and assigned hourly rate.

select p.pole,
       p.zone,
       p.area,
       r.hourly_rate
from parking_catalog.parking_schema.parking_meter_man p
inner join parking_catalog.parking_schema.rate_table r
on p.pole=r.pole;
```

## LEFT JOIN

```sql
-- returning parking meters that does not have any hourly_rate in rate table

select p.*
from parking_catalog.parking_schema.parking_meter_man p
left join parking_catalog.parking_schema.rate_table r
on p.pole=r.pole
where r.pole is null
```

## RIGHT JOIN

```sql
-- poles that no longer exist in parking table but their hourly rate is still recorded in rate table

select r.*
from parking_catalog.parking_schema.parking_meter_man p
right join parking_catalog.parking_schema.rate_table r
on p.pole=r.pole
where p.pole is null;
```

## FULL JOIN

```sql
-- display all pole details even if they have hourly rate or doesnt not exist in parking table
select p.pole as parking_pole,
       r.pole as rate_pole,
       p.zone,
       p.area,
       p.sub_area,
       r.hourly_rate,
       r.rate_type
from parking_catalog.parking_schema.parking_meter_man p
full join parking_catalog.parking_schema.rate_table r
on p.pole=r.pole
order by p.index ; --to show null vales
```

### **CSV FORMAT DATA READ**

```python
#first the dataset was loaded in the volume (parking_catlog->parking_schema->parking_volume) and then read into a dataframe.

parking_meter_df=spark.read.format("csv")\
    .option("header","true")\
    .option("inferSchema","true")\
    .load("/Volumes/parking_catalog/parking_schema/parking_volume/Parking meter locations.csv")

# display(parking_meter_df)
```

### CSV -> TEMP VIEW -> MANAGED TABLE

```python
# Creating temp view from dataset
parking_meter_df.createOrReplaceTempView("parking_meter_view")

spark.sql('''
create or replace table parking_catalog.parking_schema.parking_meter_man
as
select * from parking_meter_view''')
```
